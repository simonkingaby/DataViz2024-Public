```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.randn(50)

# Create scatter plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.randn(50) * 0.5

# Create scatter plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Non-linear Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x1 = np.random.normal(5, 1, 100)
y1 = np.random.normal(5, 1, 100)
x2 = np.random.normal(10, 1, 100)
y2 = np.random.normal(10, 1, 100)

# Create scatter plot
plt.scatter(x1, y1, label='Cluster 1', alpha=0.6)
plt.scatter(x2, y2, label='Cluster 2', alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Clustered Data')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
categories = np.random.choice(['Category 1', 'Category 2', 'Category 3'], 100)
x = np.random.rand(100)
y = np.random.rand(100)

# Create scatter plot
colors = {'Category 1': 'red', 'Category 2': 'green', 'Category 3': 'blue'}
plt.scatter(x, y, c=[colors[cat] for cat in categories])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Categorical Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000  # Bubble sizes

# Create bubble chart
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart')
plt.show()
```
